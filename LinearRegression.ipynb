```python
import numpy as np
import pandas as pd
```

```python
X = np.array([0,1,2,3,3,5,5,5,6,7,7,10])
y = np.array([96,85,82,74,95,68,76,84,58,65,75,50])
```

```python
X
```

```python
y
```

```python
x_mean = X.mean()
y_mean = y.mean()
```

```python
m = ((12*(X*y).sum())-(X.sum()*y.sum()))/((12*(X*X).sum())-(X.sum()**2))
m
```

```python
c = y_mean - m*x_mean
c
```

```python
print("Enter a value: ")
x = int(input())
solution = m*x+c
print(f"Prediction: {solution}")
```
